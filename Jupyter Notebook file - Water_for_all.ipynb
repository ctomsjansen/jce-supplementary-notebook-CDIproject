#### 1. Explanation of the Mass balance: 

The Mass balance is defined as “The change of mass of the liquid is the vessel as a function of time is equal to the amount of liquid entering vessel minus the amount of liquid leaving the vessel”. With the mass balance, you can calculate how the total mass inside the vessel changes over time. Mass cannot be destroyed nor created. 

#### 2. Solved Material Balance example: charging of a conventional CDI cell

When we have a problem in transport phenomena, we use balnce equations to solve it. The first step is to make a drawing of the problem. The second part, choose one or more balance equations, for example mass balance, energy balance, momentum balance, enz.. Then, the balance equations are solved by bookkeeping: look what comes into the volume, and what comes out of the volume. 

##### Step 1: Draw a Free Body Diagram of your system

It is a closed system with an inlet stream and an outlet stream. Here in is:

##### Step 2: Choose the balance equation
The balance equation __conservation of mass__ is always a good balance equation to start with. Because we are working with a fluid stream of liquid water and solid salt ions, we consider the so called __material balance__, which is just a mass conservation balance which also focuses on the balance of salts in the fluid streams. We will only focus on the material balance equation for the ion charging (adsorption) step. Later on, you will construct the equation for the discharging step. Ask yourself the following questions:
- Is it a steady-state or unsteady equation?
- What are the assumptions? 

General Material Balance Equation is: __Accumulation  = inlet - outlet + production - consumption__:

$$dM/dt=m_{in}-m_{out}+P-CS $$

Herein is:
- $M$ is the total mass of salt in the control volume
- $t$ is the time
- $dM/dt$ is the accumulation of salt in the control volume 
- $m_{in}$ is the mass flow in the inlet of the control volume
- $m_{out}$ is the mass flow out of the control volume
- $CS$ is consumption of salt by a chemical reaction
- $P$ is production of salt by a chemical reaction

Assumptions:
- Flowrate in is equal to flowrate out (mass conservation)
- Feed concentration is constant, $c_1$=$c_F$=$500mg/L$
- The concentration in the CDI cell depends on time (unsteady)
- The concentration in the CDI cell and the outlet are equal
- No chemical reaction, but adsorption can be assumed as a chemical reaction were salt is consumed
- Assume the consumption term is a constant, so $CS$ is independent of time

##### Step 3: Write the material balance for your system:

From the assumptions, we get the following Material Balance equation:

$$\frac{dM}{dt}=m_{in}-m_{out}-CS$$
$$\frac{d(c_{0}V_{0})}{dt}=c_{1}\phi_1-c_{2}\phi_2-CS$$

Rewriting this equation gives:

$$\frac{d(cV_{0})}{dt}=c_{F}\phi-c\phi-CS$$

##### Step 4: Solve the Material Balance. 
It is a first order linear differential equation. Tip: Use the method __Integration Factor__ to solve this differential equation. 

$$\frac{d(cV_{0})}{dt}=c_{F}\phi-c\phi-CS$$
$$V_{0}\frac{dc}{dt}=c_{F}\phi-c\phi-CS$$
$$\frac{dc}{dt}+\frac{\phi}{V_{0}}c=\frac{\phi}{V_{0}}c_{F}-\frac{CS}{V_{0}}$$

__Method of Integration Constant__
When you have an equation of the form $\frac{dy}{dx} + P(x)y = Q(x)$, 
the integration constant is calculated with $I(x)=e^{\int P(x)dx}$. 
Then, the solve differential euqation could be written as: $y(x)=\frac{1}{I(x)}(\int I(x)Q(x)dt + constant)$. 

Use the method of Integration Constant with $P(t) = \frac{\phi}{V_{0}}$ and $Q(t) = \frac{\phi}{V_{0}}c_{F}-\frac{CS}{V_{0}}$. Then, The integration constant is:
$$I(t)= e^{\int P(t)dt} = e^{\int \frac{\phi}{V_{0}}dt} = e^{\frac{\phi}{V_{0}}t}$$

Solving the differential equation then gives:
$$c(t)=\frac{1}{I(t)}(\int I(t)Q(t)dt + constant)$$
$$c(t)=e^{-\frac{\phi}{V_{0}}t}(\int e^{\frac{\phi}{V_{0}}t}(\frac{\phi}{V_{0}}c_{F}-\frac{CS}{V_{0}})dt + constant)$$
$$c(t)=e^{-\frac{\phi}{V_{0}}t}(e^{\frac{\phi}{V_{0}}t}\frac{V_{0}}{\phi}(\frac{\phi}{V_{0}}c_{F}-\frac{CS}{V_{0}})) + Ce^{-\frac{\phi}{V_{0}}t}$$
$$c(t)=c_{F}-\frac{CS}{\phi} + Ce^{-\frac{\phi}{V_{0}}t}$$

Use the Boundary Condition $c(t=0)=c_{F}$ to get the equation:
$$c(t)=\frac{CS}{\phi}(e^{-\frac{\phi}{V_{0}}t}-1)+c_{F}$$
